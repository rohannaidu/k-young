```python
#import rohan's usual tools for data-cleaning
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import scipy
from astropy.io import ascii
from astropy.table import Table, vstack
from scipy import stats
```

```python
#this of course is critical. names are alphabetical in both raw data files.
names = ['Aadit', 'Adam', 'Annette', 'Diamanta', 'Isaac', 'Isobel', 'Jay', 'Joceline', 'Keith', 'Meredith', 'Regina', 'Ross', 'Scott', 'Thaddeus']
```

```python
abroad = ascii.read("abroad.csv", names=names) #votes from study abroad
home = ascii.read("home.csv", names=names) #votes from people at yale-nus
all_data = vstack([home, abroad]) #all votes put together
```

```python
#here we get a feel for data-contamination etc. this is where I became suspicious of Hrishi et al's results.
#Regina seems to have done much better, but Keith is ranked higher in their results.
#But of course, nothing can be said till we have properly calculated K-Young.
print scipy.stats.itemfreq(all_data['Regina'])
print scipy.stats.itemfreq(all_data['Keith'])
```

```python
#to remove "No Vote","No",""(blank) and replace them with "15"(lower than the lowest possible preference)

def strip_strings(x):
    for index in range(0,len(x)):
        for second_index in range(0,len(x[0])):
            if "=>" in x[index][second_index]:
                x[index][second_index] = (x[index][second_index]).split("=>")[1].replace('"',"")
            
    for index in range(0,len(x)):
        for second_index in range(0,len(x[0])):
            if (x[index][second_index]=="NO VOTE")|(x[index][second_index]=="")|(x[index][second_index]=="NO"):
                x[index][second_index]= int(15)
            else:
                x[index][second_index]= int(x[index][second_index])
```

```python
strip_strings(all_data)
```

```python
#this is a cool library called ballotbox that I tried using. It implements Kemeny Young in a naive way and this method is STILL running
#on one of Yale's fastest supercomputers. Did this for kicks to get a sense of how much RAM brute force actually consumes:
# GREATER THAN 30 GB!!!!!!!!!!!!!!!!!!!!!!
import ballotbox
from ballotbox.ballot import BallotBox
from ballotbox.singlewinner.preferential import KemenyYoungVoting
```

```python
#here we test our understanding of the library by implementing 
#an example found at https://en.wikipedia.org/wiki/Kemeny%E2%80%93Young_method#Example

bb = BallotBox(method=KemenyYoungVoting)
preference = { "Memphis": 1, "Nashville": 2, "Chattanooga": 3, "Knoxville": 4}
bb.add_votes(preference, 84)
preference = {"Nashville": 1, "Chattanooga": 2, "Knoxville": 3, "Memphis": 4}
bb.add_votes(preference, 52)
preference = {"Knoxville": 1, "Chattanooga": 2, "Nashville": 3, "Memphis": 4}
bb.add_votes(preference, 34)
preference = {"Chattanooga": 1, "Knoxville": 2, "Nashville": 3, "Memphis": 4}
bb.add_votes(preference, 35)
bb.get_winner()
```
